```python
import pandas as pd
import seaborn as sns
import altair as alt
```

```python
apple = pd.read_csv('/Users/adamancheta/Desktop/DS4200 Project/AAPL.csv')
apple.head()
```

```python
alt.Chart(apple).mark_circle(size=60).encode(
    x=alt.X('Open', title='Open Price'),
    y=alt.Y('Volume', title='Closing Price'),
    color=alt.Color('Close', title='Close Price'),
).properties(
    title='Relationship Open Price and Close Price for Apple',
    width=700,
    height=400
)
```
